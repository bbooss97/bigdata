```python
import numpy as np
from PIL import Image
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
```

```python
#singular value decomposition with sklearn
#return the reconstructed image Ak and the energies retrieved in percentage ||Aki||^2 / ||A||^2
def svd(image,nComponents):
    svd = TruncatedSVD(n_components=nComponents)
    svd.fit(image)
    t=svd.transform(image)
    return svd.inverse_transform(t),svd.explained_variance_ratio_
```

```python
#read image in numpy array in grayscale
def readImage(filename):
    return np.array(Image.open(filename).convert('L'))
```

```python
#funtion to plot the image with matplotlib
def plotImage(image):
    plt.imshow(image, cmap='gray')
    plt.show()
```

```python
#funtion to plot the reconstructed images with and the total energy retrieved in percentage ||Ak||^2 / ||A||^2
def plotImageReconstructed(image,components):
    image,energies=svd(image,components)
    print(f"the components are: {components} and the total energy retrieved in percentage is: {energies.sum()}" )
    plotImage(image)
```

```python
image=readImage('./image.jpg')
print(f"the shape of the image is {image.shape}")
```

```python
image,energies=svd(image,200)
print(f"the total energy retrieved in percentage is: {energies.sum()}" )
```

```python
plotImage(image)
```

```python

plotImageReconstructed(image,3)
plotImageReconstructed(image,10)
plotImageReconstructed(image,20)
plotImageReconstructed(image,50)
plotImageReconstructed(image,100)
plotImageReconstructed(image,200)
plotImageReconstructed(image,300)

print("the original image is")
plotImage(readImage('./image.jpg'))
```
